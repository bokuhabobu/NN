```python
import requests

url = "https://www.ibm.com/think/topics/machine-learning"

headers = {
    "User-Agent": "Mozilla/5.0"
}

res = requests.get(url, headers=headers, timeout=10)
res.raise_for_status()

res.content[:500]
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(res.content, "html.parser")

for tag in soup(["script", "style", "nav", "header", "footer"]):
    tag.decompose()
```

```python
texts = soup.get_text(separator=" ")
texts[:1000]
```

辞書に入れるべきでない文字(\nなど）もあるので、不要な文字を消す前処理が必要。

ここら辺りは丁寧にやってほしいところ。授業ではやらないので。

```python
texts = texts.replace("\n", " ")
texts = texts.replace("\t", " ")
texts = " ".join(texts.split())

texts[:1000]
```

```python
texts = list(texts.split("."))

texts = [text.strip() for text in texts if len(text.strip()) > 20]

texts[:10]
```

```python
len(texts)
```

```python
texts = texts[:65]
```

```python
texts
```

```python
import keras

vocab_size = 5000
max_length = 120

vectorizer = keras.layers.TextVectorization(
    max_tokens=vocab_size,
    standardize="lower_and_strip_punctuation",
    split="whitespace",
    output_mode="int",
    output_sequence_length=max_length
)

vectorizer.adapt(texts)

vocabulary = vectorizer.get_vocabulary()

word_index = dict((word, index) for index, word in enumerate(vocabulary))
```

```python
word_index
```

```python
text = vectorizer([
    "machine learning models learn patterns from training data.",
    "artificial intelligence can make predictions from new data.",
    "deep learning is a type of machine learning."
]).numpy()

text
```

```python
index_word = {index: word for index, word in enumerate(vocabulary)}

def decode_text(text):  
    return " ".join(index_word.get(int(i), "?") for i in text if int(i) != 0)
```

```python
decode_text(text[0])
```
